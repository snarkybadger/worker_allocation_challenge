### Day List

```python
day_list=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
```

```python
def employee_list_comprehension(employee):
    certified_select=[d for d in employees if (d.get('type')!='Handyperson') and (d.get('type')!='Uncertified Installer')]
    uncertified_select=[d for d in employees if (d.get('type')!='Handyperson')and (d.get('type')!='Certified Installer')]
    handyperson_select=[d for d in employees if (d.get('type')!='Uncertified Installer')and (d.get('type')!='Certified Installer')]
    return(certified_select,uncertified_select,handyperson_select)
```

### Filter Type of Worker (i.e. Handyperson, Uncertified, Certified) and if they are available

```python
def employee_choose(employee,day):
    
    available_Cert=[]
    available_Uncert=[]
    available_Handy=[]
    
    for worker in employee:
        if "Certified Installer" in worker.values():
            if day in worker["days_available"]:
                available_Cert.append(worker["name"])
                
        if 'Uncertified Installer' in worker.values():
            if day in worker['days_available']:
                available_Uncert.append(worker['name'])
            
        if 'Handyperson' in worker.values():
            if day in worker['days_available']:
                available_Handy.append(worker['name'])    
    main_List=(available_Cert,available_Uncert,available_Handy)    
    return(main_List)
```

### Function 

```python
def schedule(buildings, employees):
    
    for day in day_list:
        can_do=employee_choose(employees_list,day)
        cert=can_do[0]
        uncert=can_do[1]
        handy=can_do[2]
        
        for building in buildings:
            final_list=[]
            
            if building == 'single' and len(cert)>0: #requires one certified
                final_list.append(cert[0])
                del cert[0]
                buildings_list.remove('single')    
                print(day,building,final_list)

            if building =='double' and len(cert)>0 and (len(uncert)>0 or len(handy)>0): #requires one certified AND/OR
                final_list.append(cert[0])
                del cert[0]
                
                if len(uncert)>len(cert):
                    final_list.append(uncert[0])
                    del uncert[0]
                    
                else:
                    final_list.append(cert[0])
                    del cert[0]
                buildings_list.remove('double')        
                print(day,building,final_list)
            
            if building =='commercial':
                if len(cert)>=2 and len(uncert)>=2 and (len(cert)+len(uncert)+len(handy))>=8: #requires 2 cert, 2 pending AND any 4
                    
                    final_list.append(cert[0])
                    del cert[0]
                    final_list.append(cert[0])
                    del cert[0]
                    final_list.append(uncert[0])
                    del uncert[0]
                    final_list.append(uncert[0])
                    del uncert[0]

                while len(handy)>0 and len(final_list)<=7:
                    final_list.append(handy[0])
                    del handy[0]
                    
                while len(uncert)>0 and len(final_list)<=7:
                    final_list.append(uncert[0])
                    del uncert[0]
                    
                while len(cert)>0 and len(final_list)<=7:
                    final_list.append(cert[0])
                    del cert[0]
                    
                buildings_list.remove('commercial')
                print(day,building,final_list)
                
```
